# 1. Melakukan Scraping (Halaman1-5)

```python
import requests
from bs4 import BeautifulSoup
import json
import time
import petl as etl
```

```python
BASE_URL = "https://loginmegastore.com/search"
QUERY = "kulkas"
HALAMAN = 5
HEADERS = {"User-Agent": "Mozilla/5.0"}
List_Elektronik = []
```

```python
def scrape_halaman(page_number, tampilkan_html=False):
    params = {
        "query": QUERY,
        "p": "true",
        "page": page_number
    }

    response = requests.get(BASE_URL, params=params, headers=HEADERS)
    soup = BeautifulSoup(response.text, "html.parser")

    hasil = []
    container = soup.find("div", class_="mantine-SimpleGrid-root")
    products = container.find_all("div", recursive=False)

    for i, product in enumerate(products, start=1):
        title = product.find("p", class_="mantine-focus-auto !leading-[120%] m_b6d8b162 mantine-Text-root")
        price = product.find("p", class_="mantine-focus-auto !whitespace-nowrap m_b6d8b162 mantine-Text-root")
        image = product.find("img", class_="m_9e117634 mantine-Image-root")

        if title and price and image:

            if tampilkan_html:
                print(f"\n--- Produk {i} (Page {page_number}) ---")
                print(title.prettify())
                print(price.prettify())
                print(image.prettify())
                print("-" * 40)

            hasil.append([
                title.text.strip(),
                price.text.strip(),
                image.get("src"),
                page_number
            ])

    return hasil
```

```python
for page in range(1, 6):
    scrape_halaman(page, tampilkan_html=True)
```

```python
for page in range(1, 6):
    data = scrape_halaman(page)
    print(f"\nHalaman {page}:")

    for item in data:
        print(item)
```

Loop scraping 1-5 Halaman

```python
for page in range(1, HALAMAN + 1):
    data = scrape_halaman(page)
    List_Elektronik.extend(data)
    print(f"Halaman {page}: {len(data)} data")
    time.sleep(1.5)

print("Total data:", len(List_Elektronik))
```

# 2. Menyimpan Hasil Scarping ke JSON

```python
data_json = [
    {"Title Barang": x[0], "Harga Barang": x[1], "Gambar Barang": x[2], "Page": x[3]}
    for x in List_Elektronik
]

with open("List_Elektronik.json", "w", encoding="utf-8") as f:
    json.dump(data_json, f, ensure_ascii=False, indent=4)
```

# 3. Melakukan ETL

```python
t1 = etl.fromjson("List_Elektronik.json")
t1.display(10)
print(t1)
```

```python
# Ambil hanya gambar produk (BUKAN kategori)
t2 = etl.select(t1, lambda row: "product" in row['Gambar Barang'])
print(t2)
```

```python
# Menghapus duplikat berdasarkan title + price
seen = set()
rows = []
for row in t2.dicts():
    key = (row['Title Barang'], row['Harga Barang'])
    if key not in seen:
        seen.add(key)
        rows.append(row)

t3 = etl.fromdicts(rows)
print(t3)
```

```python
# Mengubah harga barang jadi angka
t4 = etl.convert(t3, "Harga Barang", lambda x: int(x.replace("Rp", "").replace(".", "").strip()) if x else None)
print(t4)
```

```python
#Mengurutkan Pange sesuai tampilan srapping
t5 = etl.sort(t4, "Page")
print(t5)
```

# 4. Menyimpan Data Ke CSV

```python
etl.tocsv(t5, "List_Elektronik.csv")
```
